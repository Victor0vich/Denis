```python
# Преобразуем Date в datetime и устанавливаем её индексом
#data['Date'] = pd.to_datetime(data['Date'], format='%d/%m/%Y', errors='coerce')
#data.set_index('Date', inplace=True)

# Конвертируем цену закрытия в число (float)
data['Close'] = pd.to_numeric(data['Close'], errors='coerce').astype(float)

# Отфильтруем пропущенные значения
data.dropna(subset=['Close'], inplace=True)

# Берём только цену закрытия
close_prices = data['Close']

# Дневные доходности
returns = close_prices.pct_change().dropna()

# Уровень доверительности
confidence_level = 0.95

# Период удержания (например, 1 день)
holding_period = 1

def historical_var(returns, confidence_level):
    """Исторический метод VaR"""
    return returns.quantile(1 - confidence_level)

def parametric_var(returns, confidence_level):
    """Параметрический метод VaR"""
    mean_return = returns.mean()
    std_dev = returns.std()

    # Нормальная обратная функция распределения вероятности
    z_score = norm.ppf(confidence_level)

    return -(mean_return + z_score * std_dev)

# Расчёт VaR двумя методами
historical_vaR = historical_var(returns, confidence_level)
parametric_vaR = parametric_var(returns, confidence_level)

print(f"Value at Risk (Historical Method): {historical_vaR:.4%}")
print(f"Value at Risk (Parametric Method): {parametric_vaR:.4%}")
```
